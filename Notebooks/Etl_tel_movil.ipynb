```python
import pandas as pd
import funciones
import conexion_bd
import warnings
warnings.filterwarnings("ignore")
```

Traemos los conjuntos de datos par realizar las transformaciones necesarias

```python
# Datos de telefonia movil
df_ingresos_tl_movil = pd.read_excel("../Data/Telefonia_movil/Nacional/Ingresostrimestralesporlaprestaciondelserviciodetelefoniamovil_2888131699269746282.xlsx")
df_portabilidad_neta_mensual = pd.read_parquet("../Data/Telefonia_movil/Nacional/portabili_neta_mensual.parquet")
df_total_acceso_tel_movil = pd.read_excel("../Data/Telefonia_movil/Nacional/Totalnacionaldeaccesosatelefoniamovil_2888151699302990415.xlsx")
```

```python
df_ingresos_tl_movil.dtypes
```

```python
df_ingresos_tl_movil
```

Los tipos de datos de las columnas son correctos, por lo tanto agregaremos la columna id 

```python
df_ingresos_tl_movil["ingresos_tel_movil_id"] = df_ingresos_tl_movil.index
```

Eliminamos espacios y parentesis en los nombres de las columnas, pata evitar errores

```python
df_ingresos_tl_movil.columns = df_ingresos_tl_movil.columns.str.replace(" ", "_")
df_ingresos_tl_movil.columns = df_ingresos_tl_movil.columns.str.replace("(", "")
df_ingresos_tl_movil.columns = df_ingresos_tl_movil.columns.str.replace(")", "")
```

```python
# Verificamos dataset
df_ingresos_tl_movil.head()
```

Procedemos a cargar en nuestra base de datos

```python
df_ingresos_tl_movil.to_sql("ingreso_telefonia_movil", con=conexion_bd.engine, index= False, if_exists="append")
```
