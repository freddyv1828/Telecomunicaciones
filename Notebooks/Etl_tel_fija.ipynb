```python
df_ingresos_trim_tele_fija
```

Observamos que los tipos de datos son correctos, solo normalizaremos las columnas

```python
# agregamos la columna id
df_ingresos_trim_tele_fija["ingresos_id"] = df_ingresos_trim_tele_fija.index
```

```python
# Eliminamos los espacios y los parentesis de los nombres de las columnas 
df_ingresos_trim_tele_fija.columns = df_ingresos_trim_tele_fija.columns.str.replace(" ", "_") 
df_ingresos_trim_tele_fija.columns = df_ingresos_trim_tele_fija.columns.str.replace("(", "")
df_ingresos_trim_tele_fija.columns = df_ingresos_trim_tele_fija.columns.str.replace(")", "")
```

```python
df_ingresos_trim_tele_fija.to_sql("ingresos_trimestrales_telefonia_fija", con=conexion_bd.engine, index= False, if_exists="append")
```
